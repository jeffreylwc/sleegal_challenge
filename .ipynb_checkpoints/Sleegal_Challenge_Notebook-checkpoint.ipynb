```python
df = pd.concat([q3_07, q3_08, q4])
df
```

```python
df.dtypes
```

```python
df['starttime'] = pd.to_datetime(df['starttime'])
df['stoptime'] = pd.to_datetime(df['stoptime'])
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['to_station_id'].nunique()
```

```python
to_station_df = df.groupby([pd.Grouper(key='stoptime', freq='D'), "to_station_id"]).count().reset_index('stoptime')
to_station_df
```

```python
to_station_df['count'] = to_station_df['trip_id']
to_station_df = to_station_df[['stoptime', 'count']]
to_station_df
```

```python
capacity_df = capacity_df.set_index('id')
capacity_df
```

```python
merged_df = pd.merge(capacity_df, to_station_df, left_index=True, right_index =True, how='inner')
merged_df.index.name = 'id'
merged_df
```

```python
result_df = merged_df.groupby('id')['count'].agg(['mean', 'median'])
result_df
```

```python
result_df = result_df.merge(capacity_df.iloc[:, [3, 0]], left_index=True, right_index =True, how='inner')
result_df
```

```python
more_bike_docks = result_df.loc[result_df['median'] > result_df['dpcapacity']].index.to_list()
more_bike_docks_station_name = result_df.loc[result_df['median'] > result_df['dpcapacity']]['name'].to_list()
```

```python
less_bike_docks = result_df.loc[result_df['median'] < result_df['dpcapacity']].index.to_list()
less_bike_docks_station_name = result_df.loc[result_df['median'] < result_df['dpcapacity']]['name'].to_list()
```

## Results

The following two lists, "more_bike_docks" and "more_bike_docks_station_name", contain the id and name of the stations where they need to install more bike docks. I decided to use median to count the number of bikes parked at certain stations per day because median is robust while mean isn't. The stations in these lists are the ones where their medians are greater than the dpcapacity, reflecting that there are more bikes parked at the stations than their capacities allow. 

Depending on the stakeholders, we might not need to install more bike docks at certain stations. For instance, if the dpcapacity of station id 347 is 16 but its median is 15. If the stakeholders want to save money and don't mind a median of one bike not being able to park at the station, then we might not need to install more bike docks at station id 347.

The stations in the lists are the ones where the median of bikes parked at the stations are less than their capacities allow.
